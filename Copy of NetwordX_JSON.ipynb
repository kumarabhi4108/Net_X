```python
import pandas as pd
import networkx as nx
import json
pd.set_option('display.max_columns', None)
import matplotlib.pyplot as plt
import pprint
```

```python
df = pd.read_csv("https://github.com/umassdgithub/Fall-2023-DataViz/raw/main/Week-8-ForceSimulator/data/data_scopus.csv")
## drop the rows with no Author Affiliations

df= df[~pd.isnull(df['Authors with affiliations'])].copy()
```

```python
df.head(2)
```

```python
def get_first_author_affilation(x):
  try:
    return x.split(";")[0].split(",")[-1].strip()
  except Exception as ex:
    print(x)
df['First_author_Country']=df['Authors with affiliations'].apply(get_first_author_affilation)
```

```python
nodes = df['EID'].values
df = df.fillna(0)
```

```python
df['Year'].max()
```

```python
G = nx.Graph()
```

```python
nodes = []
for index, row in df.iterrows():
    author_id = row['Author(s) ID'].split(';')[0]
    country = row['First_author_Country']

    node_data = {
        'id': author_id,
        'publications': [
            {
                row['EID']: {
                    'Title': row['Title'],
                    'Year': int(row['Year']),
                    'Citations': float(row['Cited by']),
                    'Publisher': row['Publisher'],
                    'Country': country
                }
            }
        ]
    }
    nodes.append(node_data)
```

```python
len(nodes)
```

```python
nodes[0]
```

```python
edges = []
for _, row in df.iterrows():
    current_paper = row['EID']
    if ";" in row['Author(s) ID']:
        authors = row['Author(s) ID'][:-1].split(";")
        for author in authors:
            papers = df[df['Author(s) ID'].str.contains(author)]['EID'].values
            if len(papers) > 0:
                for paper in papers:
                    if paper != current_paper:
                        edges.append({"source": current_paper, "target": paper})
```

```python
# Create graph data
graph_data = {
    "directed": False,
    "multigraph": False,
    "graph": {},
    "nodes": nodes,
    "links": edges
}
```

```python
G.add_edges_from(edges)
```

```python
#pos = nx.drawing.circular_layout(G)
pos = nx.spring_layout(G)
```

```python
nx.draw(G,pos=pos, node_size=40,alpha=.2)
```

```python
# Convert to JSON and write to a file
with open('network_of_authors.json', 'w') as json_file:
    json.dump(graph_data, json_file, indent=4)
```
